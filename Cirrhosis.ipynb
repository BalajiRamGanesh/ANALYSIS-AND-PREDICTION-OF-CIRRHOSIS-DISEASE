**Data Splitting**

We can conclude that there are many characterstics of patient to be examined to get a medical outcome. we can't get a outcome by a single test and symptoms

```python
new_columns=['ID','N_Days','Status','Drug' ,'Sex','Ascites','Hepatomegaly','Spiders','Edema','Bilirubin','Cholesterol','Albumin', 'Copper', 'Alk_Phos', 'SGOT', 'Tryglicerides', 'Platelets','Prothrombin','Age1']
x=pd.get_dummies(data=cr,columns=new_columns)
y=cr['Stage']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.33,random_state=40)
```

**Feature Selection**

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder
scaler = StandardScaler()
xtrain_scaled = scaler.fit_transform(xtrain)
xtest_scaled = scaler.transform(xtest)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,accuracy_score,confusion_matrix,mean_absolute_error,mean_squared_error,classification_report
model=LinearRegression()
model.fit(xtrain,ytrain)
ypred1=model.predict(xtest)
linear_accuracy=r2_score(ytest,ypred1)
```

```python
mae = mean_absolute_error(ytest, ypred1)
mse = mean_squared_error(ytest, ypred1)
rmse = np.sqrt(mse)
print('MAE:', mae)
print('MSE:', mse)
print('RMSE:', rmse)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(xtrain, ytrain)
ypred2=dt.predict(xtest)
decision_accuracy=accuracy_score(ytest,ypred2)*100
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(xtrain, ytrain)
ypred3=rf.predict(xtest)
rf_accuracy=accuracy_score(ytest,ypred3)*100
```

```python
from sklearn.ensemble import AdaBoostClassifier
adaboost = AdaBoostClassifier()
adaboost.fit(xtrain, ytrain)
ypred4 = adaboost.predict(xtest)
ada_accuracy=accuracy_score(ytest, ypred4)*100
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gradboost = GradientBoostingClassifier()
gradboost.fit(xtrain, ytrain)
ypred5 = gradboost.predict(xtest)
grad_accuracy = accuracy_score(ytest, ypred5)*100
```

```python
from xgboost import XGBClassifier
xgboost = XGBClassifier()
xgboost.fit(xtrain, ytrain-1)
ypred6 = xgboost.predict(xtest)
xgboost_accuracy = accuracy_score(ytest, ypred6+1)*100
```

```python
from sklearn.naive_bayes import GaussianNB
naive_bayes = GaussianNB()
naive_bayes.fit(xtrain, ytrain)
ypred7 = naive_bayes.predict(xtest)
naive_accuracy = accuracy_score(ytest, ypred7)*100
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
knn = KNeighborsClassifier()
knn.fit(xtrain, ytrain)
ypred8 = knn.predict(xtest)
knn_accuracy = accuracy_score(ytest, ypred8) * 100
```

```python
from sklearn.svm import SVC
svm= SVC()
svm.fit(xtrain, ytrain)
ypred9 = svm.predict(xtest)
svm_accuracy = accuracy_score(ytest, ypred9) * 100
```

```python
print('The accuracy of linear Regression is:',linear_accuracy)
print('The accuracy of Decision Tree is:', decision_accuracy)
print('The accuracy of Random Forest is:',rf_accuracy)
print('The accuracy of AdaBoost is:',ada_accuracy)
print('The accuracy of Gradient Boosting is:',grad_accuracy)
print('The accuracy of XGBoost is:',xgboost_accuracy)
print('The accuracy of Naive Bayes is:',naive_accuracy)
print('The accuracy of support vector machines(svm) is:',svm_accuracy)
print('The accuracy of K Nearest Neigbour is:',knn_accuracy)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
encode_columns=['Status','Sex', 'Ascites','Hepatomegaly', 'Spiders', 'Edema']
for i in encode_columns:
    cr[i] =le.fit_transform(cr[i])
x=cr.drop(['Drug','ID'],axis=1)
y=le.fit_transform(cr['Drug'])
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.33,random_state=40)
gradboost = GradientBoostingClassifier()
gradboost.fit(xtrain,ytrain)
```
